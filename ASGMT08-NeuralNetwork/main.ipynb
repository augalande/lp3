```python
df=df.drop(["RowNumber","CustomerId","Surname"],axis=1)
```

```python
df=pd.get_dummies(df,columns=["Geography","Gender"])
```

```python
def min_max_normalization(name):
    global df
    df[name]=df[name].astype("float32")
    df[name]=(df[name]-df[name].min())/(df[name].max()-df[name].min())

cols=['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard','EstimatedSalary']
for col in cols:
    min_max_normalization(col)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x=df.drop("Exited",axis=1).to_numpy().astype(np.float32)
y=df["Exited"].to_numpy()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=43)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(activation="relu", units=6))#input layer
model.add(Dense(activation="relu", units=6))#hidden layer
model.add(Dense(activation="sigmoid", units=1))#output layer
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(x_train,y_train,batch_size=10,epochs=50)  
# from keras.models import Sequential
# from keras.layers import Dense

# model=Sequential()
# help("keras.Sequential.compile")
'''
Relu:- Rectified Linear Unit f(x)=max(0,x)
Sigmoid:-  f(x)= 1/(1+e^-x)
'''
```

```python
y_pred=model.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
