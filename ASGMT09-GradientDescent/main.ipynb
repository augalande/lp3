```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (x+3)**2

def derivative(x):
    return 2*(x+3)
```

```python
points=[]

def gradient_descent(starting_x,learning_rate,no_of_iterations):
    x=starting_x
    for i in range(no_of_iterations):
        points.append(x)
        print(f"Iteration :{i+1} , x= {x} , y= {function(x)}")
        grad=derivative(x)
        x=x-(grad*learning_rate)
    return x
        
```

```python
starting_x=2
learning_rate=0.1
no_of_iterations=100

min_x=gradient_descent(starting_x,learning_rate,no_of_iterations)
print(f"Local Minima at x= {min_x}")
```

```python
x_vals=np.linspace(-5,5,100)
y_vals=function(x_vals)

plt.plot(x_vals,y_vals,label="y=(x+3)**2")
plt.scatter(points,[function(point) for point in points],color="red", label="Gradient Descent Steps")
plt.title("Gradient Descent Optimisation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
